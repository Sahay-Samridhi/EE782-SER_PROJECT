```python
"""
Speech Emotion Recognition on IEMOCAP
Architecture: Transformer (Wav2Vec2)
Loss Function: Multi-task - Classification (CE) + Regression (CCC Loss)

This version uses CCC loss for continuous emotion dimensions (valence, arousal, dominance)
alongside classification loss for discrete emotion categories.
"""

# ============================================================================
# INSTALLATION
# ============================================================================
"""
!pip install kagglehub librosa soundfile matplotlib tqdm numpy pandas scikit-learn
!pip install transformers datasets torch accelerate
"""

import kagglehub
import librosa
import numpy as np
import pandas as pd
import os
import re
import warnings
from pathlib import Path
from tqdm import tqdm
from copy import deepcopy
warnings.filterwarnings('ignore')

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import Wav2Vec2Model
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, confusion_matrix, classification_report,
                            recall_score, f1_score, matthews_corrcoef, mean_squared_error)
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================================
# CONFIGURATION
# ============================================================================

class Config:
    # Dataset
    DATASET_PATH = None
    TARGET_SR = 16000
    MAX_DURATION = 10.0
    EMOTIONS = ['angry', 'happy', 'neutral', 'sad']

    # Model
    WAV2VEC2_MODEL = 'facebook/wav2vec2-base'
    FREEZE_FEATURE_ENCODER = True
    HIDDEN_SIZE = 768

    # Training
    BATCH_SIZE = 8
    LEARNING_RATE = 1e-4
    NUM_EPOCHS = 15
    EARLY_STOPPING_PATIENCE = 5
    GRADIENT_ACCUMULATION_STEPS = 4
    DROPOUT = 0.3

    # Multi-task loss weights
    LAMBDA_CLASS = 1.0  # Weight for classification loss
    LAMBDA_CCC = 1.0    # Weight for CCC regression loss

    # Data split
    TEST_SIZE = 0.15
    VAL_SIZE = 0.15

    # Device
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    RANDOM_SEED = 1337

# Set random seeds
torch.manual_seed(Config.RANDOM_SEED)
np.random.seed(Config.RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(Config.RANDOM_SEED)

# ============================================================================
# CCC LOSS FUNCTION
# ============================================================================

class CCCLoss(nn.Module):
    """
    Concordance Correlation Coefficient Loss
    Better for continuous emotion dimensions (valence, arousal, dominance)

    CCC = 2 * ρ * σ_x * σ_y / (σ_x² + σ_y² + (μ_x - μ_y)²)
    where ρ is Pearson correlation coefficient

    Loss = 1 - CCC (to minimize)
    """
    def __init__(self):
        super(CCCLoss, self).__init__()

    def forward(self, y_pred, y_true):
        """
        Args:
            y_pred: predictions (batch_size, 3) for [valence, arousal, dominance]
            y_true: ground truth (batch_size, 3)
        """
        # Mean
        y_pred_mean = torch.mean(y_pred, dim=0)
        y_true_mean = torch.mean(y_true, dim=0)

        # Variance
        y_pred_var = torch.var(y_pred, dim=0)
        y_true_var = torch.var(y_true, dim=0)

        # Covariance
        vx = y_pred - y_pred_mean
        vy = y_true - y_true_mean
        covariance = torch.sum(vx * vy, dim=0) / (y_pred.shape[0] - 1)

        # CCC for each dimension
        ccc = (2 * covariance) / (y_pred_var + y_true_var + (y_pred_mean - y_true_mean)**2 + 1e-8)

        # Average CCC across dimensions and convert to loss
        loss = 1 - torch.mean(ccc)

        return loss

# ============================================================================
# DATA LOADING
# ============================================================================

def load_iemocap_data(dataset_path):
    """
    Load IEMOCAP by parsing EmoEvaluation annotation files.
    Extract both discrete emotions AND continuous VAD (Valence, Arousal, Dominance) values.
    """
    emotion_map = {'ang': 'angry', 'hap': 'happy', 'exc': 'happy',
                   'sad': 'sad', 'neu': 'neutral'}

    # Dictionary to store utterance_id -> (emotion, valence, arousal, dominance)
    utterance_data = {}

    # Find and parse all EmoEvaluation files
    print("Parsing emotion annotation files with VAD values...")
    eval_files_found = 0

    for root, dirs, fnames in os.walk(dataset_path):
        for f in fnames:
            # Look for evaluation files in EmoEvaluation folders
            if f.endswith('.txt') and 'EmoEvaluation' in root:
                eval_files_found += 1
                eval_path = os.path.join(root, f)

                try:
                    with open(eval_path, 'r', errors='ignore') as ef:
                        for line in ef:
                            # Pattern: [start - end] utterance_id emotion [v, a, d]
                            # Example: [6.2901 - 8.2357] Ses01F_impro01_F000 neu [2.5, 2.5, 2.5]
                            match = re.match(
                                r'\[[\d\.]+ - [\d\.]+\]\s+(\S+)\s+(ang|hap|exc|sad|neu|fru|sur|fea|dis|xxx|oth)\s+\[([\d\.]+),\s*([\d\.]+),\s*([\d\.]+)\]',
                                line.strip()
                            )
                            if match:
                                utt_id = match.group(1)
                                emotion = match.group(2).lower()
                                valence = float(match.group(3))
                                arousal = float(match.group(4))
                                dominance = float(match.group(5))

                                if emotion in emotion_map:
                                    utterance_data[utt_id] = {
                                        'emotion': emotion_map[emotion],
                                        'valence': valence,
                                        'arousal': arousal,
                                        'dominance': dominance
                                    }
                except Exception as e:
                    continue

    print(f"Found {eval_files_found} evaluation files")
    print(f"Parsed {len(utterance_data)} utterance labels with VAD values")

    # Now find corresponding wav files
    print("Matching wav files with labels...")
    wav_files_dict = {}

    for root, dirs, fnames in os.walk(dataset_path):
        for f in fnames:
            if f.endswith('.wav'):
                # Get utterance ID (filename without extension)
                utt_id = f.replace('.wav', '')
                wav_files_dict[utt_id] = os.path.join(root, f)

    print(f"Found {len(wav_files_dict)} wav files")

    # Match wav files with emotions and load audio
    data = []
    print("Loading audio files...")

    for utt_id, utt_info in tqdm(utterance_data.items(), desc="Processing"):
        if utt_id not in wav_files_dict:
            continue

        if utt_info['emotion'] not in Config.EMOTIONS:
            continue

        filepath = wav_files_dict[utt_id]

        # Load audio
        try:
            audio, sr = librosa.load(filepath, sr=Config.TARGET_SR, mono=True)

            # Normalize
            audio = (audio - np.mean(audio)) / (np.std(audio) + 1e-5)

            # Trim silence
            audio, _ = librosa.effects.trim(audio, top_db=20)

            # Truncate if too long
            max_samples = int(Config.MAX_DURATION * Config.TARGET_SR)
            if len(audio) > max_samples:
                audio = audio[:max_samples]

            # Extract session ID
            session_match = re.search(r'Ses0([1-5])', filepath)
            session_id = int(session_match.group(1)) if session_match else None

            data.append({
                'filepath': filepath,
                'utterance_id': utt_id,
                'audio': audio,
                'sr': sr,
                'duration': len(audio) / sr,
                'emotion': utt_info['emotion'],
                'valence': utt_info['valence'],
                'arousal': utt_info['arousal'],
                'dominance': utt_info['dominance'],
                'session_id': session_id
            })

        except Exception as e:
            continue

    print(f"Successfully loaded {len(data)} samples")

    # Statistics
    if data:
        emotions_list = [d['emotion'] for d in data]
        unique, counts = np.unique(emotions_list, return_counts=True)
        print("\nClass distribution:")
        for emotion, count in zip(unique, counts):
            print(f"  {emotion}: {count}")

        print("\nVAD statistics:")
        print(f"  Valence:   mean={np.mean([d['valence'] for d in data]):.2f}, std={np.std([d['valence'] for d in data]):.2f}")
        print(f"  Arousal:   mean={np.mean([d['arousal'] for d in data]):.2f}, std={np.std([d['arousal'] for d in data]):.2f}")
        print(f"  Dominance: mean={np.mean([d['dominance'] for d in data]):.2f}, std={np.std([d['dominance'] for d in data]):.2f}")

    return data

# ============================================================================
# DATASET CLASS
# ============================================================================

class EmotionDataset(Dataset):
    def __init__(self, data, label_encoder):
        self.data = data
        self.label_encoder = label_encoder

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        audio = item['audio']
        label = self.label_encoder.transform([item['emotion']])[0]

        # VAD values (normalized to [0, 1] range if they're on 1-5 scale)
        vad = np.array([item['valence'], item['arousal'], item['dominance']], dtype=np.float32)
        # Normalize from [1, 5] to [0, 1]
        vad = (vad - 1.0) / 4.0

        return {
            'features': torch.FloatTensor(audio),
            'label': torch.tensor(label, dtype=torch.long),
            'vad': torch.FloatTensor(vad),
            'emotion': item['emotion']
        }

def collate_fn_wav2vec(batch):
    """Collate with padding"""
    max_len = max([item['features'].shape[0] for item in batch])

    features_list = []
    labels = []
    vads = []
    attention_masks = []

    for item in batch:
        feat = item['features']
        padding_length = max_len - feat.shape[0]

        if padding_length > 0:
            padded_feat = torch.cat([feat, torch.zeros(padding_length)])
            attention_mask = torch.cat([torch.ones(feat.shape[0]),
                                       torch.zeros(padding_length)])
        else:
            padded_feat = feat
            attention_mask = torch.ones(feat.shape[0])

        features_list.append(padded_feat)
        labels.append(item['label'])
        vads.append(item['vad'])
        attention_masks.append(attention_mask)

    return {
        'features': torch.stack(features_list),
        'labels': torch.stack(labels),
        'vads': torch.stack(vads),
        'attention_mask': torch.stack(attention_masks)
    }

# ============================================================================
# MODEL (Multi-task: Classification + VAD Regression)
# ============================================================================

class Wav2Vec2ForEmotionRecognition(nn.Module):
    """
    Wav2vec2 for multi-task emotion recognition:
    1. Discrete emotion classification (4 classes)
    2. Continuous VAD regression (3 dimensions)
    """

    def __init__(self, num_labels):
        super().__init__()

        self.wav2vec2 = Wav2Vec2Model.from_pretrained(Config.WAV2VEC2_MODEL)

        if Config.FREEZE_FEATURE_ENCODER:
            self.wav2vec2.feature_extractor.eval()
            for param in self.wav2vec2.feature_extractor.parameters():
                param.requires_grad = False

        hidden_size = self.wav2vec2.config.hidden_size

        # Shared representation
        self.shared_layer = nn.Sequential(
            nn.Dropout(Config.DROPOUT),
            nn.Linear(hidden_size, hidden_size),
            nn.Tanh(),
            nn.Dropout(Config.DROPOUT)
        )

        # Classification head
        self.classifier = nn.Linear(hidden_size, num_labels)

        # VAD regression head
        self.vad_regressor = nn.Linear(hidden_size, 3)  # valence, arousal, dominance

    def _get_feat_extract_output_lengths(self, input_lengths):
        def _conv_out_length(input_length, kernel_size, stride):
            return (input_length - kernel_size) // stride + 1

        for kernel_size, stride in zip([10, 3, 3, 3, 3, 2, 2], [5, 2, 2, 2, 2, 2, 2]):
            input_lengths = _conv_out_length(input_lengths, kernel_size, stride)

        return input_lengths

    def forward(self, input_values, attention_mask=None):
        if Config.FREEZE_FEATURE_ENCODER:
            self.wav2vec2.feature_extractor.eval()

        outputs = self.wav2vec2(
            input_values,
            attention_mask=attention_mask,
            output_hidden_states=True
        )

        hidden_states = outputs.last_hidden_state

        # Mean pooling with mask
        if attention_mask is not None:
            batch_size = hidden_states.shape[0]
            input_lengths = attention_mask.sum(-1)
            output_lengths = self._get_feat_extract_output_lengths(input_lengths)

            max_len = hidden_states.shape[1]
            mask = torch.arange(max_len, device=hidden_states.device)[None, :] < output_lengths[:, None]
            mask = mask.unsqueeze(-1).expand(hidden_states.size()).float()

            masked_hidden = hidden_states * mask
            sum_hidden = masked_hidden.sum(dim=1)
            sum_mask = mask.sum(dim=1)
            mean_hidden = sum_hidden / (sum_mask + 1e-9)
        else:
            mean_hidden = hidden_states.mean(dim=1)

        # Shared representation
        shared_repr = self.shared_layer(mean_hidden)

        # Classification logits
        logits = self.classifier(shared_repr)

        # VAD predictions (apply sigmoid to bound to [0, 1])
        vad_pred = torch.sigmoid(self.vad_regressor(shared_repr))

        return logits, vad_pred

# ============================================================================
# TRAINING
# ============================================================================

def calculate_class_weights(data, label_encoder):
    """Calculate class weights"""
    emotions = [d['emotion'] for d in data]
    labels = label_encoder.transform(emotions)

    class_counts = np.bincount(labels)
    total_samples = len(labels)
    n_classes = len(class_counts)

    class_weights = total_samples / (n_classes * class_counts)
    class_weights = torch.FloatTensor(class_weights).to(Config.DEVICE)

    print("\nClass weights:")
    for i, emotion in enumerate(Config.EMOTIONS):
        print(f"  {emotion}: {class_weights[i]:.4f} (count: {class_counts[i]})")

    return class_weights

def train_epoch(model, dataloader, optimizer, criterion_class, criterion_ccc, device):
    """Train one epoch with multi-task loss"""
    model.train()

    if Config.FREEZE_FEATURE_ENCODER:
        model.wav2vec2.feature_extractor.eval()

    total_loss = 0
    total_class_loss = 0
    total_ccc_loss = 0
    predictions = []
    true_labels = []

    optimizer.zero_grad()
    pbar = tqdm(dataloader, desc="Training", leave=False)

    for batch_idx, batch in enumerate(pbar):
        features = batch['features'].to(device)
        labels = batch['labels'].to(device)
        vads = batch['vads'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        # Forward pass
        logits, vad_pred = model(features, attention_mask=attention_mask)

        # Multi-task loss
        loss_class = criterion_class(logits, labels)
        loss_ccc = criterion_ccc(vad_pred, vads)
        loss = Config.LAMBDA_CLASS * loss_class + Config.LAMBDA_CCC * loss_ccc

        loss = loss / Config.GRADIENT_ACCUMULATION_STEPS
        loss.backward()

        if (batch_idx + 1) % Config.GRADIENT_ACCUMULATION_STEPS == 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            optimizer.zero_grad()

        total_loss += loss.item() * Config.GRADIENT_ACCUMULATION_STEPS
        total_class_loss += loss_class.item()
        total_ccc_loss += loss_ccc.item()

        preds = torch.argmax(logits, dim=1)
        predictions.extend(preds.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

        pbar.set_postfix({
            'loss': f'{total_loss/(batch_idx+1):.4f}',
            'cls': f'{total_class_loss/(batch_idx+1):.4f}',
            'ccc': f'{total_ccc_loss/(batch_idx+1):.4f}',
            'acc': f'{accuracy_score(true_labels, predictions):.4f}'
        })

    avg_loss = total_loss / len(dataloader)
    avg_class_loss = total_class_loss / len(dataloader)
    avg_ccc_loss = total_ccc_loss / len(dataloader)
    accuracy = accuracy_score(true_labels, predictions)
    uar = recall_score(true_labels, predictions, average='macro')

    return avg_loss, avg_class_loss, avg_ccc_loss, accuracy, uar

def evaluate(model, dataloader, criterion_class, criterion_ccc, device):
    """Evaluate model with multi-task metrics"""
    model.eval()
    total_loss = 0
    total_class_loss = 0
    total_ccc_loss = 0
    predictions = []
    true_labels = []
    all_vad_pred = []
    all_vad_true = []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            features = batch['features'].to(device)
            labels = batch['labels'].to(device)
            vads = batch['vads'].to(device)
            attention_mask = batch['attention_mask'].to(device)

            logits, vad_pred = model(features, attention_mask=attention_mask)

            loss_class = criterion_class(logits, labels)
            loss_ccc = criterion_ccc(vad_pred, vads)
            loss = Config.LAMBDA_CLASS * loss_class + Config.LAMBDA_CCC * loss_ccc

            total_loss += loss.item()
            total_class_loss += loss_class.item()
            total_ccc_loss += loss_ccc.item()

            preds = torch.argmax(logits, dim=1)
            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

            all_vad_pred.append(vad_pred.cpu().numpy())
            all_vad_true.append(vads.cpu().numpy())

    avg_loss = total_loss / len(dataloader)
    avg_class_loss = total_class_loss / len(dataloader)
    avg_ccc_loss = total_ccc_loss / len(dataloader)
    accuracy = accuracy_score(true_labels, predictions)
    uar = recall_score(true_labels, predictions, average='macro')

    # Calculate VAD MSE
    all_vad_pred = np.vstack(all_vad_pred)
    all_vad_true = np.vstack(all_vad_true)
    vad_mse = mean_squared_error(all_vad_true, all_vad_pred)

    return avg_loss, avg_class_loss, avg_ccc_loss, accuracy, uar, vad_mse, predictions, true_labels

# ============================================================================
# MAIN TRAINING FUNCTION
# ============================================================================

def train_model(data, label_encoder):
    """Main training function with multi-task learning"""
    print("\n" + "="*80)
    print("Training Wav2Vec2 with Multi-Task Learning")
    print("Classification Loss: Cross Entropy | Regression Loss: CCC")
    print("="*80)

    # Calculate class weights
    class_weights = calculate_class_weights(data, label_encoder)

    # Split data
    train_val_data, test_data = train_test_split(
        data,
        test_size=Config.TEST_SIZE,
        stratify=[d['emotion'] for d in data],
        random_state=Config.RANDOM_SEED
    )

    train_data, val_data = train_test_split(
        train_val_data,
        test_size=Config.VAL_SIZE / (1 - Config.TEST_SIZE),
        stratify=[d['emotion'] for d in train_val_data],
        random_state=Config.RANDOM_SEED
    )

    print(f"\nData split:")
    print(f"  Train: {len(train_data)} ({len(train_data)/len(data)*100:.1f}%)")
    print(f"  Val:   {len(val_data)} ({len(val_data)/len(data)*100:.1f}%)")
    print(f"  Test:  {len(test_data)} ({len(test_data)/len(data)*100:.1f}%)")

    # Create datasets
    train_dataset = EmotionDataset(train_data, label_encoder)
    val_dataset = EmotionDataset(val_data, label_encoder)
    test_dataset = EmotionDataset(test_data, label_encoder)

    # Weighted sampler for balanced training
    train_labels = label_encoder.transform([d['emotion'] for d in train_data])
    class_counts = np.bincount(train_labels)
    sample_weights = (1.0 / class_counts)[train_labels]

    sampler = torch.utils.data.WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(sample_weights),
        replacement=True
    )

    # Dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=Config.BATCH_SIZE,
        sampler=sampler,
        collate_fn=collate_fn_wav2vec,
        num_workers=2
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=Config.BATCH_SIZE,
        shuffle=False,
        collate_fn=collate_fn_wav2vec,
        num_workers=2
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=Config.BATCH_SIZE,
        shuffle=False,
        collate_fn=collate_fn_wav2vec,
        num_workers=2
    )

    # Model, optimizer, losses
    model = Wav2Vec2ForEmotionRecognition(len(Config.EMOTIONS)).to(Config.DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=Config.LEARNING_RATE, weight_decay=0.01)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)

    criterion_class = nn.CrossEntropyLoss(weight=class_weights)
    criterion_ccc = CCCLoss()

    # Training loop
    best_val_uar = 0
    best_epoch = 0
    patience_counter = 0
    best_model_state = None

    history = {
        'train_loss': [], 'train_class_loss': [], 'train_ccc_loss': [],
        'train_acc': [], 'train_uar': [],
        'val_loss': [], 'val_class_loss': [], 'val_ccc_loss': [],
        'val_acc': [], 'val_uar': [], 'val_vad_mse': []
    }

    for epoch in range(Config.NUM_EPOCHS):
        print(f"\nEpoch {epoch+1}/{Config.NUM_EPOCHS}")
        print("-" * 60)

        train_loss, train_class_loss, train_ccc_loss, train_acc, train_uar = train_epoch(
            model, train_loader, optimizer, criterion_class, criterion_ccc, Config.DEVICE
        )

        val_loss, val_class_loss, val_ccc_loss, val_acc, val_uar, val_vad_mse, val_preds, val_labels = evaluate(
            model, val_loader, criterion_class, criterion_ccc, Config.DEVICE
        )

        scheduler.step(val_uar)

        history['train_loss'].append(train_loss)
        history['train_class_loss'].append(train_class_loss)
        history['train_ccc_loss'].append(train_ccc_loss)
        history['train_acc'].append(train_acc)
        history['train_uar'].append(train_uar)
        history['val_loss'].append(val_loss)
        history['val_class_loss'].append(val_class_loss)
        history['val_ccc_loss'].append(val_ccc_loss)
        history['val_acc'].append(val_acc)
        history['val_uar'].append(val_uar)
        history['val_vad_mse'].append(val_vad_mse)

        print(f"TRAIN → Loss: {train_loss:.4f} (Class: {train_class_loss:.4f}, CCC: {train_ccc_loss:.4f}) | Acc: {train_acc:.4f} | UAR: {train_uar:.4f}")
        print(f"VAL   → Loss: {val_loss:.4f} (Class: {val_class_loss:.4f}, CCC: {val_ccc_loss:.4f}) | Acc: {val_acc:.4f} | UAR: {val_uar:.4f} | VAD MSE: {val_vad_mse:.4f}")

        if val_uar > best_val_uar:
            best_val_uar = val_uar
            best_epoch = epoch
            best_model_state = deepcopy(model.state_dict())
            patience_counter = 0
            print(f"✅ NEW BEST! Val UAR: {best_val_uar:.4f}")
        else:
            patience_counter += 1
            print(f"⏳ Patience: {patience_counter}/{Config.EARLY_STOPPING_PATIENCE}")

        if patience_counter >= Config.EARLY_STOPPING_PATIENCE:
            print(f"\n🛑 Early stopping at epoch {epoch+1}")
            break

    # Load best model and evaluate on test
    model.load_state_dict(best_model_state)
    test_loss, test_class_loss, test_ccc_loss, test_acc, test_uar, test_vad_mse, test_preds, test_labels = evaluate(
        model, test_loader, criterion_class, criterion_ccc, Config.DEVICE
    )

    # Final metrics
    f1 = f1_score(test_labels, test_preds, average='macro')
    mcc = matthews_corrcoef(test_labels, test_preds)

    print(f"\n{'='*80}")
    print("FINAL RESULTS")
    print(f"{'='*80}")
    print(f"Best model from Epoch {best_epoch+1}")
    print(f"\nClassification Metrics:")
    print(f"  Test UAR: {test_uar:.4f}")
    print(f"  Test Accuracy: {test_acc:.4f}")
    print(f"  Test F1-Macro: {f1:.4f}")
    print(f"  Test MCC: {mcc:.4f}")
    print(f"\nRegression Metrics:")
    print(f"  Test VAD MSE: {test_vad_mse:.4f}")
    print(f"  Test CCC Loss: {test_ccc_loss:.4f}")

    print("\nClassification Report:")
    print(classification_report(test_labels, test_preds, target_names=Config.EMOTIONS, digits=4))

    # Confusion matrix
    cm = confusion_matrix(test_labels, test_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=Config.EMOTIONS, yticklabels=Config.EMOTIONS)
    plt.title(f'Confusion Matrix - Wav2Vec2 Multi-Task (CE + CCC)\nUAR: {test_uar:.4f}')
    plt.ylabel('True')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig('wav2vec2_ccc_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Plot training history
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # Total Loss
    axes[0, 0].plot(history['train_loss'], label='Train Loss', linewidth=2)
    axes[0, 0].plot(history['val_loss'], label='Val Loss', linewidth=2)
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Total Loss')
    axes[0, 0].set_title('Total Loss (Classification + CCC)')
    axes[0, 0].legend()
    axes[0, 0].grid(alpha=0.3)

    # UAR
    axes[0, 1].plot(history['train_uar'], label='Train UAR', linewidth=2)
    axes[0, 1].plot(history['val_uar'], label='Val UAR', linewidth=2)
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('UAR')
    axes[0, 1].set_title('Unweighted Average Recall (Classification)')
    axes[0, 1].legend()
    axes[0, 1].grid(alpha=0.3)

    # Classification Loss
    axes[1, 0].plot(history['train_class_loss'], label='Train CE Loss', linewidth=2)
    axes[1, 0].plot(history['val_class_loss'], label='Val CE Loss', linewidth=2)
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Cross Entropy Loss')
    axes[1, 0].set_title('Classification Loss (Cross Entropy)')
    axes[1, 0].legend()
    axes[1, 0].grid(alpha=0.3)

    # CCC Loss
    axes[1, 1].plot(history['train_ccc_loss'], label='Train CCC Loss', linewidth=2)
    axes[1, 1].plot(history['val_ccc_loss'], label='Val CCC Loss', linewidth=2)
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('CCC Loss')
    axes[1, 1].set_title('Regression Loss (CCC for VAD)')
    axes[1, 1].legend()
    axes[1, 1].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig('wav2vec2_ccc_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Additional plot for VAD MSE
    plt.figure(figsize=(10, 5))
    plt.plot(history['val_vad_mse'], label='Val VAD MSE', linewidth=2, color='green')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Squared Error')
    plt.title('VAD Prediction Error (Valence, Arousal, Dominance)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig('wav2vec2_vad_mse.png', dpi=300, bbox_inches='tight')
    plt.show()

    return model, history

# ============================================================================
# RUN TRAINING
# ============================================================================

if __name__ == "__main__":
    # Download dataset
    print("Downloading IEMOCAP dataset...")
    path = kagglehub.dataset_download("dejolilandry/iemocapfullrelease")
    Config.DATASET_PATH = path
    print(f"Dataset path: {path}")

    # Load data using the enhanced loader with VAD
    print("\n" + "="*80)
    print("Loading IEMOCAP Dataset with VAD Annotations")
    print("="*80)
    data = load_iemocap_data(path)

    if not data:
        print("ERROR: No data loaded!")
        print("\nTroubleshooting:")
        print("1. Check if EmoEvaluation folders exist in the dataset")
        print("2. Verify .txt annotation files are present")
        print("3. Ensure .wav files are in the correct structure")
        print("4. Check if VAD values [v, a, d] are in annotation files")
    else:
        # Create DataFrame for statistics
        df = pd.DataFrame(data)
        print("\n" + "="*80)
        print("Dataset Statistics")
        print("="*80)
        print(f"Total samples: {len(df)}")
        print(f"\nEmotion distribution:")
        print(df['emotion'].value_counts())
        print(f"\nAverage duration: {df['duration'].mean():.2f}s")
        print(f"Duration range: [{df['duration'].min():.2f}s, {df['duration'].max():.2f}s]")

        # VAD correlation with emotions
        print("\nVAD values by emotion:")
        for emotion in Config.EMOTIONS:
            emotion_data = df[df['emotion'] == emotion]
            print(f"\n{emotion.upper()}:")
            print(f"  Valence:   {emotion_data['valence'].mean():.2f} ± {emotion_data['valence'].std():.2f}")
            print(f"  Arousal:   {emotion_data['arousal'].mean():.2f} ± {emotion_data['arousal'].std():.2f}")
            print(f"  Dominance: {emotion_data['dominance'].mean():.2f} ± {emotion_data['dominance'].std():.2f}")

        # Prepare labels
        label_encoder = LabelEncoder()
        label_encoder.fit(Config.EMOTIONS)

        # Train
        model, history = train_model(data, label_encoder)

        # Save model
        torch.save({
            'model_state_dict': model.state_dict(),
            'config': Config,
            'label_encoder': label_encoder,
            'history': history
        }, 'wav2vec2_ccc_emotion_model.pth')

        print("\n✅ Training complete!")
        print("✅ Model saved to 'wav2vec2_ccc_emotion_model.pth'")
        print("\n" + "="*80)
        print("MODEL ARCHITECTURE: Multi-Task Learning")
        print("  → Task 1: Emotion Classification (4 classes) - Cross Entropy Loss")
        print("  → Task 2: VAD Regression (3 dimensions) - CCC Loss")
        print("="*80)
```
